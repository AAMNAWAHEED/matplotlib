```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x=np.arange(1,100)
y=np.sin(x)
```

```python
plt.plot(x,y,color='pink',linestyle='dotted')
plt.xlabel('numbers')
plt.ylabel('sin(x)')
plt.grid(True,which='both')
plt.minorticks_on()
```

```python
plt.plot(x,y)
plt.xlim(1,20)
plt.ylim(-0.50,0.50)
plt.legend(['green','blue'],loc='lower right')
```
